```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("Employee Attrition.csv")   # Column: Satisfaction
```

```python
print(df.columns)
```

```python
# Count responses
counts = df["satisfaction_level"].value_counts()
```

```python
print(counts)
```

```python
# Bar chart
counts.plot(kind='bar')
plt.xlabel("Satisfaction Level")
plt.ylabel("Response Count")
plt.title("Survey Satisfaction Results")
plt.show()
```
